```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Student_Marks.csv')
df.head()
```

```python
#Mean
import pandas as pd
mean1 = df['Marks'].mean()
print(mean1)
```

```python
#Max
import pandas as pd
max1 = df['Marks'].max()
print(max1)
```

```python
#Min
import pandas as pd
min1 = df['Marks'].min()
print(min1)
```

```python
#Median
import pandas as pd
median1 = df['Marks'].median()
print(median1)
```

```python
#Standard deviation
import pandas as pd
std1 = df['Marks'].std()
print(std1)
```

```python
#Varience
import pandas as pd
var1 = df['Marks'].var()
print(var1)
```

```python
#Mode
import pandas as pd
mode1 = df['number_courses'].mode()
print(mode1)
```

```python
#Sorting
import pandas as pd
sorted_df = df.sort_values(by=['Marks'], ascending=True)
print(sorted_df)
```

```python
#Count
import pandas as pd
count1 = df['Marks'].count()
print(count1)
```

```python
#Filtering
import pandas as pd
df.filter(["number_courses"])
```

```python
#Grouping
import pandas as pd
gk = df.groupby('number_courses')
gk.first()
```

```python
#Aggregate
import pandas as pd
df.aggregate(['sum', 'min'])
```
